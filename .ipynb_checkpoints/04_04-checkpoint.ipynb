```python
import pandas as pd
```

```python
states_babies = pd.read_csv("state_baby_names.csv")
```

```python
states_babies["Year"] == 2014
```

```python
states_2014_babies = states_babies.loc[states_babies["Year"] == 2014 , : ]
```

```python
ca_babies = states_2014_babies.loc[states_2014_babies["State"] == "CA"]
```

```python
sorted_ca_babies = ca_babies.sort_values("Count",ascending = False)
```
