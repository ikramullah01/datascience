```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listings = pd.read_csv("Airbnb_NYC_2019.csv")
```

```python
listings
```

```python
sn.countplot(x = "neighbourhood_group", data = listings)
plt.show()
```

```python
sn.barplot(x = "neighbourhood_group", y = "price", data = listings , errorbar=None)
plt.show()
```
